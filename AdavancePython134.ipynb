```python
entered_std = set()
rejected_std = set()

n = int(input("Enter Entry attempt: "))

count = 0
while count < n:
    name = input("Enter student names: ").strip().lower()

    clean_name = name.replace(" ", "")

    if clean_name not in entered_std:
        print(name, "Entry allowed")
        entered_std.add(clean_name)
    else:
        print(name, "already entered. entry rejected")
        rejected_std.add(clean_name)

    count += 1

print("Event details:")

print("\nTotal unique entry :", len(entered_std))
for student in entered_std:
    print(student)

print("\nTotal rejected entry :", len(rejected_std))
for student in rejected_std:
    print(student)

```

ROOM ALLOTMENT USING DICTIONARY

```python

rooms = {i: None for i in range(100, 110)}

print("Available rooms :")
print(*rooms.keys())

attempts = int(input("Enter Entry attempt: "))

count = 0
while count < attempts:
    rmno = int(input("Enter the room number : "))

    if rmno not in rooms:
        print("Invalid room number !!")
    elif rooms.get(rmno):
        print("Room already alloted !!")
    else:
        guest = input("Enter your name : ")
        rooms[rmno] = guest
        print(f"Room Alloted successfully to {guest}")

    count += 1

delrom = int(input("Enter the room to vacate : "))

if delrom in rooms and rooms[delrom] is not None:
    rooms[delrom] = None
    print("Room checkout Successfully !!")
else:
    print("Cannot vacate Room or room already vacate")

print("All Booked rooms details :")
for room, guest in rooms.items():
    if guest is not None:
        print(f"{room} : {guest}")

print("All Vacate rooms details :")
for room, guest in rooms.items():
    if guest is None:
        print(f"{room} : {guest}")

print("Thank You !!!")
```

ATM PROGRAM WITH FUNCTION

```python
def login(pin, correct_pin=1674):
    return pin == correct_pin


def deposit(balance, transactions):
    amt = int(input("Enter the amount for deposit : "))
    balance += amt
    transactions.append(f"Deposit of {amt}")
    print(f"Deposit Successful of {amt}")
    return balance


def withdraw(balance, transactions):
    amt = int(input("Enter the amount you want to withdraw : "))

    if amt > balance:
        print("Insufficient Balance")
    else:
        balance -= amt
        transactions.append(f"Withdraw of {amt}")
        print(f"Withdraw Successful of {amt}")

    return balance


def history(transactions):
    if not transactions:
        print("No Transactions")
    else:
        for t in transactions:
            print(t)


def view_balance(balance):
    print(f"Account balance : {balance}")

balance = 0
transactions = []
approve = False

for cnt in range(3):
    print(f"Available chances : {3 - cnt}")
    code = int(input("Enter your pin : "))

    if login(code):
        approve = True
        break
    else:
        print("Wrong Pin")

while approve:
    print("\n1 for Deposit")
    print("2 for Withdraw")
    print("3 for Balance Inquiry")
    print("4 for All Transactions")
    print("5 for Exit\n")

    ch = int(input("Enter Your Choice : "))

    if ch == 1:
        balance = deposit(balance, transactions)

    elif ch == 2:
        balance = withdraw(balance, transactions)

    elif ch == 3:
        view_balance(balance)

    elif ch == 4:
        history(transactions)

    elif ch == 5:
        print("Exiting...!")
        break

    else:
        print("Invalid Choice !!")

print("Thank you for using")
```

LIBRARY MANAGEMENT

```python
library = []

def add_book():
    book = input("Enter book name: ")
    library.append(book)
    print("Book added successfully!\n")

def view_books():
    if len(library) == 0:
        print("Library is empty.\n")
    else:
        print("Available Books:")
        for i, book in enumerate(library, 1):
            print(i, book)
        print()

def issue_book():
    book = input("Enter book name to issue: ")
    if book in library:
        library.remove(book)
        print("Book issued successfully!\n")
    else:
        print("Book not available.\n")

def return_book():
    book = input("Enter book name to return: ")
    library.append(book)
    print("Book returned successfully!\n")

while True:
    print("------ LIBRARY MENU ------")
    print("1.Add Book")
    print("2.View Books")
    print("3.Issue Book")
    print("4.Return Book")
    print("5.Exit")

    choice = input("Enter choice: ")

    if choice == "1":
        add_book()
    elif choice == "2":
        view_books()
    elif choice == "3":
        issue_book()
    elif choice == "4":
        return_book()
    elif choice == "5":
        print("Thank you!")
        break
    else:
        print("Invalid choice!\n")
```

```python
class Vehicle:
    def __init__(self, number, owner, phone):
        self.number = number
        self.owner = owner
        self.phone = phone


class ParkingSystem:
    def __init__(self):
        self.slots = {i: {"status": False, "vehicle": None, "fee": 5, "vnum": None} for i in range(1, 11)}
        self.vehicles_inside = set()
        self.history = {}
        self.total_money = 0
        self.his_id = 1

    def is_free(self, slot):
        return not self.slots[slot]["status"]

    def allot_slot(self, slot, vehicle):
        self.slots[slot]["vehicle"] = vehicle
        self.slots[slot]["status"] = True
        self.slots[slot]["vnum"] = vehicle.number
        self.vehicles_inside.add(vehicle.number)

        print(f"{vehicle.number} allotted to slot {slot} successfully")

    def checkout(self, slot):
        if self.is_free(slot):
            print("Slot already empty !!")
            return

        data = self.slots[slot]
        vehicle = data["vehicle"]

        self.history[self.his_id] = {"slot": slot, "vehicle": vehicle}
        self.his_id += 1

        self.vehicles_inside.remove(vehicle.number)
        self.total_money += data["fee"]

        self.slots[slot] = {"status": False, "vehicle": None, "fee": 5, "vnum": None}

        print(f"Slot {slot} is now empty")

    def show_allotted(self):
        print("Allotted Slots:")
        for s, info in self.slots.items():
            if info["status"]:
                print(f"Slot {s} allotted to vehicle : {info['vnum']}")

    def show_total(self):
        print("Total earned:", self.total_money)


print("Welcome to Parking Lot System")
print("Slots Available: 1 - 10")

parking = ParkingSystem()

bookings = int(input("Enter number of vehicle bookings: "))

for i in range(bookings):
    slot = int(input("Enter slot number: "))

    if slot not in parking.slots:
        print("Invalid slot number!")
        continue

    if parking.is_free(slot):
        vnum = input("Enter vehicle number: ")

        if vnum in parking.vehicles_inside:
            print("Vehicle already parked!")
            continue

        owner = input("Enter owner name: ")
        phone = input("Enter phone number: ")

        v = Vehicle(vnum, owner, phone)
        parking.allot_slot(slot, v)

    else:
        print("Slot already occupied!")

checkouts = int(input("Enter number of checkouts: "))

for i in range(checkouts):
    slot = int(input("Enter slot number to checkout: "))

    if slot in parking.slots:
        parking.checkout(slot)
    else:
        print("Invalid slot number!")

print()
parking.show_allotted()

print()
parking.show_total()
```

```python
class TaskManager:
    def __init__(self):
        self.task_list = {}
        self.task_id = 1

    def add_task(self):
        task = input("Enter your task: ")
        deadline = input("Enter deadline (dd/mm/yyyy): ")
        priority = int(input("Enter priority: "))
        tag = input("Enter task tag (Pending): ")

        self.task_list[self.task_id] = {
            "task": task,
            "deadline": deadline,
            "priority": priority,
            "status": tag
        }

        self.task_id += 1

    def update_status(self, tid):
        if tid not in self.task_list:
            print("Task ID not found !!")
            return

        new_status = input("Enter the status of the task: ")
        self.task_list[tid]["status"] = new_status

    def filter_tasks(self, status):
        print(f"Tasks filtered according to {status}")
        for i, data in self.task_list.items():
            if data["status"] == status:
                print(f"TaskID: {i}, Task: {data['task']}, Deadline: {data['deadline']}")

    def sort_priority(self):
        sorted_tasks = sorted(self.task_list.items(),
                              key=lambda x: x[1]["priority"])

        for tid, task in sorted_tasks:
            print(tid, task)

    def show_tasks(self):
        for i, data in self.task_list.items():
            print(f"TaskID: {i}, Task: {data['task']}, Deadline: {data['deadline']}")


tm = TaskManager()

n = int(input("Enter the number of tasks to enter: "))
for i in range(n):
    tm.add_task()

print("All the tasks are:")
tm.show_tasks()

n2 = int(input("Enter the number of tasks to change the status: "))
for j in range(n2):
    tid = int(input("Enter the task ID: "))
    tm.update_status(tid)

print("View tasks by tag:")
tag = input("Enter tag to sort: ")
tm.filter_tasks(tag)

print("View Task based on priority:")
tm.sort_priority()
```

```python
class Shop:
    def __init__(self):
        self.items = {}
        self.item_id = 1

    def add_item(self):
        name = input("Enter product name: ")
        stock = int(input("Enter stock quantity: "))
        price = float(input("Enter product price: "))
        discount = int(input("Enter discount percentage: "))

        final_price = price - (discount / 100) * price

        self.items[self.item_id] = {
            "name": name,
            "stock": stock,
            "price": price,
            "discount": discount,
            "final_price": final_price
        }

        self.item_id += 1

    def update_discount(self):
        pid = int(input("Enter product ID: "))

        if pid not in self.items:
            print("Product not found !!")
            return

        new_disc = int(input("Enter new discount: "))
        price = self.items[pid]["price"]

        final_price = price - (new_disc / 100) * price

        self.items[pid]["discount"] = new_disc
        self.items[pid]["final_price"] = final_price

        print("Discount updated successfully !!")

    def show_products(self):
        for pid, data in self.items.items():
            print(f"ProductID: {pid}, Product: {data['name']}, Stock: {data['stock']}, Price: {data['price']}, Discount Price: {data['final_price']}")

    def purchase(self):
        print("Available products:")
        self.show_products()

        cart = []
        total_bill = 0

        n = int(input("Enter number of items to purchase: "))

        for i in range(n):
            pid = int(input("Enter product ID: "))

            if pid not in self.items:
                print("Invalid Product ID")
                continue

            qty = int(input("Enter quantity: "))

            if qty > self.items[pid]["stock"]:
                print("Quantity exceeds stock")
                continue

            price = self.items[pid]["final_price"]
            cost = price * qty

            self.items[pid]["stock"] -= qty
            cart.append((pid, qty, cost))

            total_bill += cost

        print("Cart Details:")
        for c in cart:
            print("ProductID:", c[0], "Quantity:", c[1], "Cost:", c[2])

        print("Total Bill:", total_bill)

    def sort_price(self):
        sorted_items = sorted(self.items.items(), key=lambda x: x[1]["final_price"])

        for pid, item in sorted_items:
            print(pid, item)


shop = Shop()

n = int(input("Enter the number of products to add: "))
for i in range(n):
    shop.add_item()

print("All the products are:")
shop.show_products()

n2 = int(input("Enter number of products to change discount: "))
for i in range(n2):
    shop.update_discount()

print("Products after discount update:")
shop.show_products()

print("Make Purchase:")
shop.purchase()

print("Products sorted by price:")
shop.sort_price()
```

```python
class Cinema:

    def __init__(self):
        self.movie = "MS DHONI"
        self.seat_status = {i: "Free" for i in range(1, 21)}
        self.ticket_price = 100

    def show_seats(self):
        print("Seat Status for", self.movie)
        for seat, status in self.seat_status.items():
            print("Seat", seat, ":", status)

    def choose_seats(self):
        self.show_seats()
        n = int(input("Enter number of seats to book: "))
        chosen = []

        for i in range(n):
            seat = int(input("Enter seat number: "))

            if seat not in self.seat_status:
                print("Invalid seat number")
                continue

            if self.seat_status[seat] == "Booked":
                print("Seat already booked")
                continue

            chosen.append(seat)

        return chosen

    def reserve_ticket(self):
        seats = self.choose_seats()

        if len(seats) == 0:
            print("No seats selected")
            return

        for s in seats:
            self.seat_status[s] = "Booked"

        self.show_ticket(seats)

    def show_ticket(self, seats):
        print("Ticket Information:")
        print("Movie:", self.movie)
        print("Seats Booked:", seats)
        print("Total Tickets:", len(seats))
        print("Total Price:", len(seats) * self.ticket_price)
        print("Booking Successful!")


c = Cinema()

while True:
    print("\n1. View Seats")
    print("2. Book Ticket")
    print("3. Exit")

    choice = int(input("Enter your choice: "))

    if choice == 1:
        c.show_seats()
    elif choice == 2:
        c.reserve_ticket()
    elif choice == 3:
        print("Thank you!")
        break
    else:
        print("Invalid choice")
```

```python
class StudentResult:

    def __init__(self):
        self.sem1_marks = {i: 0 for i in range(1, 6)}
        self.sem2_marks = {i: 0 for i in range(1, 6)}
        self.cgpa = {i: "NA" for i in range(1, 6)}
        self.percent = {i: 0 for i in range(1, 6)}

    def view_sem1(self):
        print("Python")
        for r, sgpa in self.sem1_marks.items():
            print(f"Roll {r} SGPA {sgpa}")

    def view_sem2(self):
        print("JAVA")
        for r, sgpa in self.sem2_marks.items():
            print(f"Roll {r} SGPA {sgpa}")

    def enter_marks(self):
        print("Enter SGPA Details:")
        for r in range(1, 6):
            self.sem1_marks[r] = float(input(f"Enter Sem1 SGPA of {r}: "))
            self.sem2_marks[r] = float(input(f"Enter Sem2 SGPA of {r}: "))
        self.compute_cgpa()

    def compute_cgpa(self):
        for r in range(1, 6):
            avg = round((self.sem1_marks[r] + self.sem2_marks[r]) / 2, 3)
            self.cgpa[r] = avg
            self.percent[r] = avg * 9.5

    def display_cgpa(self):
        print("CGPA")
        for r, val in self.cgpa.items():
            print(f"Roll {r} CGPA {val}")

    def display_percentage(self):
        print("Percentage")
        for r, val in self.percent.items():
            print(f"Roll {r} Percentage {val}")


sr = StudentResult()

while True:
    print("\n1. Check Sem1 SGPA")
    print("2. Check Sem2 SGPA")
    print("3. Enter SGPA")
    print("4. Show CGPA")
    print("5. Show Percentage")
    print("6. Exit")

    choice = int(input("Enter choice: "))

    if choice == 1:
        sr.view_sem1()
    elif choice == 2:
        sr.view_sem2()
    elif choice == 3:
        sr.enter_marks()
    elif choice == 4:
        sr.display_cgpa()
    elif choice == 5:
        sr.display_percentage()
    elif choice == 6:
        print("Thank you!")
        break
    else:
        print("Invalid Choice")
```

```python
class Inventory:

    def __init__(self):
        self.item_id = 0
        self.items = {}

    def add_item(self):
        name = input("Enter product name: ")
        quantity = int(input("Enter quantity: "))
        price = float(input("Enter price per piece: "))

        self.items[self.item_id] = {
            "name": name,
            "qty": quantity,
            "price": price
        }

        print("Product added successfully with ID:", self.item_id)
        self.item_id += 1

    def add_many_items(self):
        n = int(input("Enter number of products: "))
        for i in range(n):
            self.add_item()

    def show_items(self):
        if len(self.items) == 0:
            print("No products available")
        else:
            print("Product Details")
            for pid, data in self.items.items():
                print(f"ID:{pid} Name:{data['name']} Stock:{data['qty']} Price:{data['price']}")

    def issue_item(self):
        pid = int(input("Enter product ID: "))

        if pid not in self.items:
            print("Invalid Product ID")
            return

        qty = int(input("Enter quantity to allot: "))

        if qty <= self.items[pid]["qty"]:
            self.items[pid]["qty"] -= qty
            total_cost = qty * self.items[pid]["price"]

            print("Product allotted")
            print("Total price:", total_cost)
        else:
            print("Not enough stock")


inv = Inventory()

while True:
    print("\n1. Add a Product")
    print("2. Add Multiple Products")
    print("3. Check All Products")
    print("4. Allot Product")
    print("5. Exit")

    choice = int(input("Enter choice: "))

    if choice == 1:
        inv.add_item()
    elif choice == 2:
        inv.add_many_items()
    elif choice == 3:
        inv.show_items()
    elif choice == 4:
        inv.issue_item()
    elif choice == 5:
        print("Thank you")
        break
    else:
        print("Invalid choice")
```

```python
class CredentialManager:

    def __init__(self):
        self.cid = 0
        self.data = {}

    def add_credential(self):
        website = input("Enter site name: ")
        user = input("Enter username or email: ")
        pwd = input("Enter password: ")

        self.data[self.cid] = {
            "site": website,
            "user": user,
            "pwd": pwd
        }

        print("Password added successfully with ID:", self.cid)
        self.cid += 1

    def add_multiple(self):
        n = int(input("Enter number of passwords: "))
        for i in range(n):
            self.add_credential()

    def show_all(self):
        if len(self.data) == 0:
            print("No passwords stored")
        else:
            print("Stored Passwords")
            for i, info in self.data.items():
                print(f"ID:{i} Site:{info['site']} Username/Email:{info['user']} Password:{info['pwd']}")

    def retrieve_password(self):
        cid = int(input("Enter ID for password: "))

        if cid in self.data:
            info = self.data[cid]
            print("Site:", info["site"])
            print("Username or email:", info["user"])
            print("Password:", info["pwd"])
        else:
            print("Not a valid ID")

    def update_password(self):
        cid = int(input("Enter ID to edit password: "))

        if cid in self.data:
            new_pwd = input("Enter new password: ")
            self.data[cid]["pwd"] = new_pwd
            print("Password updated successfully")
        else:
            print("Invalid ID")


cm = CredentialManager()

while True:
    print("\n1. Add a Password")
    print("2. Add Multiple Passwords")
    print("3. Check All Passwords")
    print("4. Get Password")
    print("5. Edit Password")
    print("6. Exit")

    choice = int(input("Enter choice: "))

    if choice == 1:
        cm.add_credential()
    elif choice == 2:
        cm.add_multiple()
    elif choice == 3:
        cm.show_all()
    elif choice == 4:
        cm.retrieve_password()
    elif choice == 5:
        cm.update_password()
    elif choice == 6:
        print("Thank you")
        break
    else:
        print("Invalid choice")
```

```python
class CustomerManager:

    def __init__(self):
        self.customer_id = 0
        self.log_id = 0
        self.customer_data = {}
        self.communication_logs = {}

    def add_customer(self):
        name = input("Enter customer name: ")
        email = input("Enter email: ")
        phone = input("Enter phone: ")

        self.customer_data[self.customer_id] = {
            "name": name,
            "email": email,
            "phone": phone
        }

        print("Customer added with ID:", self.customer_id)
        self.customer_id += 1

    def add_multiple_customers(self):
        n = int(input("Enter number of customers: "))
        for i in range(n):
            self.add_customer()

    def show_customers(self):
        if len(self.customer_data) == 0:
            print("No customers available")
        else:
            for cid, info in self.customer_data.items():
                print(f"ID:{cid} Name:{info['name']} Email:{info['email']} Phone:{info['phone']}")

    def add_note(self):
        cid = int(input("Enter customer ID: "))

        if cid in self.customer_data:
            note = input("Enter communication note: ")

            self.communication_logs[self.log_id] = {
                "cid": cid,
                "note": note
            }

            print("Log added with ID:", self.log_id)
            self.log_id += 1
        else:
            print("Invalid customer ID")

    def show_logs(self):
        cid = int(input("Enter customer ID to view logs: "))
        found = False

        for lid, log in self.communication_logs.items():
            if log["cid"] == cid:
                print(f"LogID:{lid} Message:{log['note']}")
                found = True

        if not found:
            print("No logs found")


cm = CustomerManager()

while True:
    print("\n1. Add Customer")
    print("2. Add Multiple Customers")
    print("3. View Customers")
    print("4. Add Communication Log")
    print("5. View Logs")
    print("6. Exit")

    choice = int(input("Enter choice: "))

    if choice == 1:
        cm.add_customer()
    elif choice == 2:
        cm.add_multiple_customers()
    elif choice == 3:
        cm.show_customers()
    elif choice == 4:
        cm.add_note()
    elif choice == 5:
        cm.show_logs()
    elif choice == 6:
        print("Thank you")
        break
    else:
        print("Invalid choice")
```
